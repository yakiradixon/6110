# Normal distribution

```python
# Import matplotlib and numpy packages
# Import distributions from scipy.stats
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import t
```

```python
# Let X~Normal(10,2) #mean 10, sdev = 2

# Define some values of X to use to graph the density curve
xvals = np.linspace(norm.ppf(0.0001, 10, 2), norm.ppf(0.9999, 10, 2), 100)

# norm.ppf(0.0001, 10, 2) finds the value of x where the lower 0.01% of the normal
# distribution (mean=10, std =2)  -->basically lower bound

# np.linspace(start, stop, num =10)
# Generate 100 evenly spaced values between start and stop


# Graph the density curve for Normal(10,2)
plt.plot(xvals, norm.pdf(x=xvals, loc=10, scale=2))

plt.ylabel('Density', fontsize=12)
plt.xlabel("X", fontsize=12)
plt.show()
```

```python
# Let X~Normal(10,2)
# Calculate the probability of less than a value, P(X<=8), using cdf()

#cdf --> Cumulative distribution function

norm.cdf(x=8, loc=10, scale=2)
```

```python
# Let X~Normal(10,2)

# Calculate the probability of greater than a value, P(X>8)=1-P(X<=8), using cdf()
1 - norm.cdf(x=8, loc=10, scale=2)
```

```python
# Let X~Normal(10,2)

# Calculate the probability between two values, P(8<X<12), using cdf()
norm.cdf(x=12, loc=10, scale=2) - norm.cdf(x=8, loc=10, scale=2)
```

### Standard Normal distirbution (Z-distribution)

```python
#How about the stanard normal distribution function?
#The probability of X < 1 in the standard normal distribution
#For the standard normal distribution, loc = 0 (mean is 0) and scale = 1 (std =1)
#In other words, P(Z < 1)

prob = norm.cdf (x =1, loc = 0, scale = 1)

print (prob)
```

```python
# Let X~t(4)

# Define some values of X to use to graph the density curve
xvals = np.linspace(t.ppf(0.001, df=4), t.ppf(0.999, df=4), 100)

# Graph the density curve for t(4)
# t.pdf --> probability distribution function for the Student's t-distribution

plt.plot(xvals, t.pdf(x=xvals, df=4))
plt.ylabel('Density', fontsize=12)
plt.xlabel("X", fontsize=12)
plt.show()
```

```python
# Let X~t(4)

# Calculate P(t <=0)
# t.cdf -> cuumlative distribution function for the student's t distribution

t.cdf(x=0, df=4)
```

```python
# Let X~t(4)

# Calculate probability in the tails P(t < -2 or t > 2)
t.cdf(x=-2, df=4) + (1 - t.cdf(x=2, df=4))
```

```python
# Let X~t(4)

# Using the symmetry of the t-distribution curve, calculate P(t < -2 or t > 2)
t.cdf(x=-2, df=4) * 2
```
