### Row Subsetting using boolean expression



```python
# column subsetting
print(myData['Income']) # returns a series (1 D array)
print(myData.Income) # . notation
print(myData[['Income']]) #returns a dataframe (2 D array)
```

```python
#row subsetting using boolean expression

myData['Income'] > 90000 #boolean expression -> returns a list boolean values (tures/falses)
```

```python
myData[ myData['Income'] > 90000  ]
```

```python
(myData[ myData['Income'] > 90000  ]).shape
```

```python
#high_spending_cust = (myData['Spending2020'] > 800 ) &  (myData['Spending2021'] > 800)

high_spending_cust = myData[ (myData['Spending2020'] > 1000) &  (myData['Spending2021'] > 1000)  ]

# & --> and operator,   | --> or operator


high_spending_cust
```

```python
high_spending_cust2 = myData[ (myData['Income'] > 90000) |  (myData['Spending2020'] > 2000)  ]
high_spending_cust2
```
