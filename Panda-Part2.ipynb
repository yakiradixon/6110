## What : means in slices [start:end]

start: where to begin (inclusive)

stop: where to end

.iloc (position-based): end is exclusive
e.g., df.iloc[0:3] → positions 0,1,2

.loc (label-based): end is inclusive
e.g., df.loc["A":"C"] → labels A, B, and C



```python
# Example code to show iloc vs loc
import pandas as pd
df = pd.DataFrame({
    "math": [85, 90, 78, 92],
    "english": [88, 76, 95, 80],
    "science": [91, 89, 84, 77],
    "history": [70, 85, 88, 93]
}, index=["s1", "s2", "s3", "s4"])

df
```

```python
df.loc["s2":"s3"]
```

```python
df.iloc[1:3]  #same as loc["s2":"s3"] --> return 2nd row (1 st position) and 3rd row (2nd position)
```

```python
#If you want to replace the missing value with the mean of the service value
# 2. Replace missing values with the mean of the column 'Service'
# Calculate the mean of the 'Service' column
mean_service = df['Service'].mean()

mean_service
```
