```python
import numpy as np
from scipy.stats import t

```

## The quiz problem (one tail test), att-ch9-2

```python
mu_0 = 12 #hypothesis testing

x_bar = 14 #sample mean

s = 4.27 #sample standard devation
n = 25   #sample size

alpha = 0.05  # level of sig alpha = 5%

df = n - 1 #degree of freedom
```

### First step: find the t_stat

```python
t_stat = (x_bar-mu_0)/(s/np.sqrt(n))
t_stat
```

### 2nd Step: Calculate the p-value

```python
p_value = 1-t.cdf(t_stat,df)   #t.cdf returns the probability (area to the left)
p_value
```

```python
if p_value < alpha:
  print("Reject the null as we have strong evidence againt the null")
else:
  print("Fail to reject the null")
```
